### t-SNE Embeddings

```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from sklearn.manifold import TSNE
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Device:", device)
```

```python
dataset = load_dataset('surrey-nlp/BESSTIE-CW-26')
```

```python
test_df = pd.DataFrame(dataset['test'])
test_df['Sarcasm'] = test_df['Sarcasm'].astype(int)
```

```python
# load roberta
tokenizer = AutoTokenizer.from_pretrained('roberta-base')
model = AutoModelForSequenceClassification.from_pretrained('models/roberta_sarcasm')
model = model.to(device)
model.eval()

print("Model loaded!")
print("Test size:", len(test_df))
```

```python
def get_embeddings(texts, batch_size=32):
    all_embeddings = []

    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i+batch_size]
        encodings = tokenizer(
            batch_texts,
            truncation=True,
            padding=True,
            max_length=128,
            return_tensors='pt'
        )
        
        input_ids = encodings['input_ids'].to(device)
        attention_mask = encodings['attention_mask'].to(device)

        with torch.no_grad():
            # Access the RoBERTa encoder directly to get hidden states
            # We bypass the classification head as we only need the representations
            outputs = model.roberta(
                input_ids=input_ids,
                attention_mask=attention_mask
            )
            '''
            Take the CLS token's embedding (the first position). 
            This token is designed to capture the whole sentence's meaning, 
            and it's the standard choice for classification in BERT‑style models
            '''
            cls_embedding = outputs.last_hidden_state[:, 0, :]
            all_embeddings.append(cls_embedding.cpu().numpy())
    return np.vstack(all_embeddings) # Stack all batch embeddings into one array of shape (num_samples, 768)

# Extract embeddings for the full test set
embeddings = get_embeddings(test_df['text'].tolist())
print("Embeddings shape: ", embeddings.shape)
```

```python
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
embeddings_2d = tsne.fit_transform(embeddings)
print(f"t-SNE done!")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1 - coloured by variety
colors_variety = {'en-UK': '#1f77b4', 'en-AU': '#ff7f0e', 'en-IN': '#2ca02c'}
for variety, color in colors_variety.items():
    mask = test_df['variety'] == variety
    axes[0].scatter(
        embeddings_2d[mask, 0],
        embeddings_2d[mask, 1],
        c=color, label=variety, alpha=0.5, s=10
    )
axes[0].set_title('t-SNE - Coloured by Variety')
axes[0].legend()

# Plot 2 - coloured by sarcasm label
colors_sarcasm = {0: '#4575b4', 1: '#d73027'}
labels_sarcasm = {0: 'Not Sarcastic', 1: 'Sarcastic'}
for label, color in colors_sarcasm.items():
    mask = test_df['Sarcasm'] == label
    axes[1].scatter(
        embeddings_2d[mask, 0],
        embeddings_2d[mask, 1],
        c=color, label=labels_sarcasm[label], alpha=0.5, s=10
    )
axes[1].set_title('t-SNE - Coloured by Sarcasm Label')
axes[1].legend()

plt.suptitle('RoBERTa Embedding Space Visualisation', fontsize=13)
plt.tight_layout()
plt.savefig('figures/tsne_embeddings.png', dpi=150)
plt.show()
print("Done!")
```
