````python
import subprocess
import sys
import os

def run_notebook(path):
    print(f"Running {path}...")
    result = subprocess.run(
        [sys.executable, "-m", "jupyter", "nbconvert",
         "--to", "notebook", "--execute", "--inplace",
         "--ExecutePreprocessor.timeout=3600", path],
        capture_output=True, text=True
    )
    if result.returncode == 0:
        print(f"  Done: {path}")
    else:
        print(f"  Failed: {path}")
        print(result.stderr[-300:])

print("Main entry point initialised.")
print("Run each section below individually or run all at once.")
````
